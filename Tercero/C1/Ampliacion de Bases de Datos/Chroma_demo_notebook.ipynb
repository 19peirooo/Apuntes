### 6.2.3. Primer arranque: cliente en memoria vs. persistente

- **En memoria** (para pruebas rГЎpidas): los datos se pierden al cerrar el proceso.
- **Persistente**: guarda el Г­ndice en una carpeta y lo reaprovecha en siguientes ejecuciones.

**Ejemplo mГ­nimo (persistente) con colecciГіn, inserciГіn y consulta:**


````python
import chromadb
from chromadb.utils import embedding_functions

# Ruta local donde se guardarГЎ el Г­ndice
client = chromadb.PersistentClient(path="./chroma_data")

# OpciГіn A: embeddings locales (sin API)
ef = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)

# Crear (o recuperar) la colecciГіn con mГ©trica de coseno
col = client.get_or_create_collection(
    name="demo_docs",
    metadata={"hnsw:space": "cosine"},
    embedding_function=ef
)

# Insertar documentos con metadatos e IDs
docs = [
    "GuГ­a rГЎpida de instalaciГіn de Python",
    "Buenas prГЎcticas de control de versiones con Git",
    "IntroducciГіn a la bГєsqueda semГЎntica"
]
metas = [
    {"tema": "python", "nivel": "bГЎsico"},
    {"tema": "git", "nivel": "bГЎsico"},
    {"tema": "vectores", "nivel": "medio"}
]
ids = ["doc1", "doc2", "doc3"]
col.add(documents=docs, metadatas=metas, ids=ids)

# Consulta por texto (Chroma genera el embedding y busca por similitud)
res = col.query(query_texts=["ВїCГіmo empezar con Git?"], n_results=2)
print(res["ids"], res["metadatas"])
````

**Puntos clave del ejemplo**

- `PersistentClient(path=...)` crea/usa la carpeta `./chroma_data` para el Г­ndice.
- `embedding_function` permite que Chroma convierta automГЎticamente cada texto a vector (si no se pasa, se pueden aГ±adir vectores ya calculados vГ­a `embeddings=`).
- Los **metadatos** (*payload*) se guardan junto al documento y se usan luego para filtrar.


### 6.2.4. Filtros por metadatos y upserts rГЎpidos


````python
# AГ±adir filtros en la consulta
res = col.query(
    query_texts=["instalaciГіn"],
    n_results=3,
    where={"tema": "python"}  # filtro por payload
)
print(res["ids"], res["metadatas"])

# Actualizar o reinsertar (mismo id в†’ upsert)
col.update(
    ids=["doc1"],
    documents=["InstalaciГіn de Python en macOS y Windows (actualizada)"],
    metadatas=[{"tema": "python", "nivel": "bГЎsico"}]
)

# Eliminar
col.delete(ids=["doc3"])
````

### 6.2.5. Estructura de carpetas y persistencia

- La carpeta de persistencia (por ejemplo, `./chroma_data/`) contendrГЎ los ficheros del Г­ndice y tablas auxiliares (formato interno basado en DuckDB/Parquet).
- Es seguro copiarla o respaldarla como parte del proyecto; al volver a crear el `PersistentClient` con la misma ruta, la colecciГіn queda disponible.


### 6.2.6. Uso con embeddings externos (APIs)

Si ya existen vectores (por ejemplo, generados con una API), se pueden insertar directamente **sin** `embedding_function`.


````python
import chromadb
client = chromadb.PersistentClient(path="./chroma_data")
col_pre = client.get_or_create_collection(name="precomputed", metadata={"hnsw:space":"cosine"})

# embeddings: lista de listas (floats), longitud = nВє documentos, dimensiГіn fija
col_pre.add(
    ids=["x1", "x2"],
    documents=["Texto A", "Texto B"],  # opcional si solo quieres buscar por vector
    embeddings=[[0.1, 0.2, 0.05], [0.05, -0.1, 0.2]],
    metadatas=[{"tag":"A"}, {"tag":"B"}]
)
````

> RecomendaciГіn: normalizar los vectores si se va a usar **coseno** o **inner product** con vectores unitarios.


### 6.2.7. Comprobaciones y problemas habituales

- *вЂњNo encuentra sentence-transformersвЂќ*: instalar con `pip install sentence-transformers` o, si no se quieren dependencias, aГ±adir `embeddings=` precomputados.
- *вЂњResultados inconsistentesвЂќ*: mezclar documentos insertados con y sin normalizaciГіn de vectores; usar siempre la misma mГ©trica (`{"hnsw:space": "cosine"}`, `"l2"`, `"ip"`) en la colecciГіn.
- *вЂњCarpeta crece muchoвЂќ*: eliminar documentos que ya no hagan falta y realizar backups/limpieza periГіdica de la carpeta de persistencia.


## 6.3. Colecciones, documentos y filtros

GuГ­a, paso a paso, para crear una colecciГіn con persistencia en disco, ingerir documentos con metadatos y consultar por similitud con y sin filtros. Dos formas habituales de generar embeddings: local con **sentence-transformers** y mediante API (OpenAI).


### 6.3.1. PreparaciГіn rГЎpida

- Entorno: Python 3.10+ (recomendado), `chromadb` instalado y, si se desea generar embeddings locales, `sentence-transformers`.
- Carpeta de persistencia: `./chroma_data` (se puede cambiar).
- MГ©trica por colecciГіn: usaremos **cosine** (lo mГЎs comГєn para embeddings normalizados).


````python
# 0) Cliente persistente y funciГіn de embeddings local
import chromadb
from chromadb.utils import embedding_functions

client = chromadb.PersistentClient(path="./chroma_data")
ef_local = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"  # 384D; rГЎpido y suficiente para prГЎcticas
)

# 1) Crear (o recuperar) colecciГіn con mГ©trica coseno
col = client.get_or_create_collection(
    name="curso_demo",
    metadata={"hnsw:space": "cosine"},
    embedding_function=ef_local
)
````

### 6.3.2. Dataset de ejemplo (documentos + metadatos)


````python
docs = [
    "InstalaciГіn de Python en Windows y macOS",
    "Primeros pasos con Git y buenas prГЎcticas de commit",
    "Arrays y operaciones vectorizadas con NumPy",
    "VisualizaciГіn de datos con Matplotlib: lГ­neas y barras",
    "Limpieza de datos: valores nulos y outliers",
    "BГєsqueda semГЎntica con embeddings y bases vectoriales",
    "Flujo bГЎsico de RAG: recuperar, filtrar y reenrutar pasajes",
    "Control de versiones avanzado: ramas, merge y resoluciГіn de conflictos",
    "IntroducciГіn a pandas: Series y DataFrames",
    "HiperparГЎmetros en modelos clГЎsicos de ML",
    "Transformers y embeddings de texto para bГєsqueda",
    "MГ©tricas de evaluaciГіn: recall@k, precisiГіn y MRR"
]
metas = [
    {"tema":"python", "nivel":"basico", "idioma":"es", "anio":2024},
    {"tema":"git", "nivel":"basico", "idioma":"es", "anio":2024},
    {"tema":"numpy", "nivel":"medio", "idioma":"es", "anio":2024},
    {"tema":"viz", "nivel":"basico", "idioma":"es", "anio":2024},
    {"tema":"datos", "nivel":"medio", "idioma":"es", "anio":2024},
    {"tema":"vectores", "nivel":"medio", "idioma":"es", "anio":2024},
    {"tema":"rag", "nivel":"medio", "idioma":"es", "anio":2024},
    {"tema":"git", "nivel":"avanzado", "idioma":"es", "anio":2023},
    {"tema":"pandas", "nivel":"basico", "idioma":"es", "anio":2023},
    {"tema":"ml", "nivel":"medio", "idioma":"es", "anio":2022},
    {"tema":"vectores", "nivel":"medio", "idioma":"es", "anio":2023},
    {"tema":"eval", "nivel":"basico", "idioma":"es", "anio":2022}
]
ids = [f"d{i+1}" for i in range(len(docs))]

col.add(documents=docs, metadatas=metas, ids=ids)
print("Ingesta completada:", len(ids), "documentos")
````

### 6.3.3. Consultas bГЎsicas (texto в†’ vector в†’ NN) y filtrado simple


````python
# Top-3 mГЎs cercanos a una intenciГіn
q1 = col.query(query_texts=["ВїCГіmo crear y fusionar ramas con Git?"], n_results=3)
print(q1["ids"], q1["metadatas"])

# Con filtro: solo documentos de 'git'
q2 = col.query(query_texts=["merge de ramas"], n_results=3, where={"tema":"git"})
print(q2["ids"], q2["metadatas"])

# Con filtro combinado (AND): 'vectores' en 2023
q3 = col.query(query_texts=["bГєsqueda semГЎntica"], n_results=3, where= {"$and": [
            {"tema": "vectores"},
            {"anio": 2023},
        ]})

print(q3["ids"], q3["metadatas"])
````

### 6.3.5. Persistencia: cerrar, reabrir y consultar

Al usar `PersistentClient`, el Г­ndice queda en disco y puede reutilizarse desde otro script o una nueva sesiГіn.


````python
# En otro script/sesiГіn (simulado aquГ­):
import chromadb
from chromadb.utils import embedding_functions

client2 = chromadb.PersistentClient(path="./chroma_data")
ef_local2 = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)
col2 = client2.get_or_create_collection(
    name="curso_demo", metadata={"hnsw:space":"cosine"}, embedding_function=ef_local2
)

res = col2.query(query_texts=["instalaciГіn de Python"], n_results=2)
print(res["ids"], res["metadatas"])
````
